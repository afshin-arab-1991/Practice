```python
game = []
m_dict = {'POINTER': f"\n‚Ä¢‚Ä¢‚Ä¢ ",
          'WSIZE':"What size of Tic Tac Toe you want to play? ",
          'INVALID1':"Please enter a valid number between 2 and 10 exclusive; ",
          'INVALID2':"You did not entered a number. Try again. Error:\n",
          'INVALID3':"Invalid Input. Error:\n",
          'RWCOL':"\nPlease type the row and column number respectively seperated ba a space. ",
          'EXIT':"‚Ä¢‚Ä¢‚Ä¢ Game Exits ‚Ä¢‚Ä¢‚Ä¢\n",
          'RESET':"‚Ä¢‚Ä¢‚Ä¢ Restrating‚Ä¢‚Ä¢‚Ä¢\n",
          'IDXERROR1':f"The numbers should be in the range of the game board, between 0 to ",
          'IDXERROR2': "inclusive.\n",
          'VALERROR':"Please enter numbers.\n",
          'BLANK':'',
          'GAMEOVER':"The game is over, would you like to play again? (y/n) ",
          'BYE':"Byeeeee!",
          'NOWINNER':"The game is over with no winner!!! Would you like to play again? (y/n) ",
          'TAKEN': "This position is already taken! Try another.",
          'WINNERIS':"The winner is player",
          'HELP':"After this point you can enter q to exit and r to restart the game.\n"}
####
pointer = m_dict['POINTER']
blank = m_dict['BLANK']
tip = pointer + m_dict['HELP']
greetings = 'Hello, ' + os.getlogin() + '! Welcome to the TIC-TAC-TOE game by Afshin. üòÉ'
```

```python
def anime_print(string,delay):
    for i in string:
        time.sleep(delay)
        print(i,end="")
```

```python
def letsplay(game_map ,player=0, row=0, column=0, just_display = False):
    try:
        if game_map[row][column] != 0:
            print(pointer,m_dict['TAKEN'])
            return game_map, False
        if not just_display:  #it means if just_display is false run the code below;
            game_map[row][column] = player
            print(" "*3 + "  ".join([str(i) for i in range(len(game_map))]))
        for count,row in enumerate(game_map):
            print(count,row)
        return game_map, True
        
        if  just_display:  #it means if just_display is false run the code below;
            print(" "*3 + "  ".join([str(i) for i in range(len(game_map))]))
        for count,row in enumerate(game_map):
            print(count,row)
        return game_map, False

    except Exception as e:
        print(pointer,m_dict['INVALID3'],e,"\n",m_dict['IDXERROR1'],len(game)-1,m_dict['IDXERROR2'])
        return game_map, False
    
    except IndexError as e:
        print(pointer,m_dict['INVALID3'],e,"\n",m_dict['IDXERROR1'],len(game)-1,m_dict['IDXERROR2'])
        return game_map, False
```

```python
def win(current_game):
    win_type = ''
    player_no = ''
    iswinner = False
    
    def check_numbers(no):
        if no.count(no[0]) == len(no) and no[0] != 0:
            iswinner = True
            return True
        else:
            return False          
    
    def show_message(w,p):
        deco_u, deco_l, deco_e = "‚ïî"+"‚ïê"*65+"‚ïó"+"\n" , "‚ïö"+"‚ïê"*65+"‚ïù"+"\n" , " "*11+"‚ïë"
        message = f"{deco_u}‚ïë\t {m_dict['WINNERIS']} {player_no}, on {win_type}{deco_e}\n{deco_l}"
        print (message)
        
    
    #########horizontal winner:
    for row in current_game:
        if check_numbers(row):
            win_type, player_no = "horizontal (‚Üî) axis" , f"{row[0]}"
            show_message(win_type,player_no)
            return True

    #########vertical winner:
    for col in range(len(current_game)):
        check = []
        for row in current_game:
            check.append(row[col])
        if check_numbers(check):
            win_type, player_no = "vertical (‚Üï) axis  " , f"{check[0]}"
            show_message(win_type,player_no)
            return True

    ########diagonal 1 winner :
    check = []
    for row in range(len(current_game)):
        check.append(current_game[row][row])
    if check_numbers(check):
        win_type, player_no = "diagonal (\\) axis  ", f"{check[0]}"
        show_message(win_type,player_no)
        return True
    
    ########diagonal 2 winner :
    diagonal = []
    for row, col in enumerate(reversed(range(len(current_game)))):
        diagonal.append(current_game[row][col])
    if check_numbers(diagonal):
        win_type, player_no = "diagonal (/) axis  " , f"{diagonal[0]}"
        show_message(win_type,player_no)
        return True 
    
    return False
```

```python
play= True
size = 0
z_sum = 999
tdelay = 0.2
edelay = 1
print(greetings)
while play:
    while size <= 2 or size >=10:
        try:
            size = int(input(m_dict['WSIZE']))
            if int(size) <= 2 or int(size) >=10:
                print(pointer,m_dict['INVALID1'])
                valid_size = False
            else:
                board  = [[0 for j in range(size)]for i in range(size)]
                valid_size = True
                print(tip)
        except ValueError as e:
            print(pointer,m_dict['INVALID2'], e ,"\n",m_dict['INVALID1'])
    game_won = False
    game, _ = letsplay(board, just_display=True)
    player_choice = itertools.cycle([1,2])
    while not game_won:
        current_player = next(player_choice)
        print(pointer,"Player "+str(current_player)+"'s turn;")
        played = False
        while not played:
        #############
                value = str(input(m_dict['RWCOL'])).rstrip().lstrip()
                if value.lower() == "q":
                    anime_print(m_dict['EXIT'],tdelay)
                    time.sleep(edelay)
                    play, played, game_won = False, True, True
                elif value.lower() == "r":
                    anime_print(m_dict['RESET'],tdelay)
                    time.sleep(edelay)
                    board = []
                    played, game_won, size = True, True, False
                else:
                    try:
                        row_choice = int(value.split(" ")[0])
                        col_choice = int(value.split(" ")[1])
                        game, played = letsplay(game, current_player, row_choice, col_choice)
                        z_sum = sum(x == 0 for r in game for x in r) 
                    except ValueError as e:
                        print(pointer,m_dict['INVALID3'],e,"\n",m_dict['VALERROR'],
                              m_dict['IDXERROR1'],len(game)-1,m_dict['IDXERROR2'])
                        played = False                    
        #############
        if win(game):
            game_won = True
            again  = input(m_dict['GAMEOVER'])
            if again.lower() == "y":
                anime_print(m_dict['RESET'],tdelay)
                time.sleep(edelay)
                board, size = [],False
            elif again.lower() == "n":
                anime_print(m_dict['BYE'],tdelay)
                time.sleep(edelay)
                play = False
            else:
                print(pointer,m_dict['INVALID3'])
                anime_print(m_dict['EXIT'],tdelay)
                time.sleep(1)
                play = False
        
        if z_sum == 0 and not game_won:
            game_won = True
            again  = input(m_dict['NOWINNER'])
            if again.lower() == "y":
                anime_print(m_dict['RESET'],tdelay)
                time.sleep(edelay)
                board, size = [],False
            elif again.lower() == "n":
                anime_print(m_dict['BYE'],tdelay)
                time.sleep(edelay)
                play = False
            else:
                print(pointer,m_dict['INVALID3'])
                anime_print(m_dict['EXIT'],tdelay)
                time.sleep(edelay)
                play = False
```
